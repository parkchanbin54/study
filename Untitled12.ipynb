```python
#count people by range of bmi parameter (float) bmi rate
def calculateSB(bm):
    bmiset=np.zeros(4)
    for i in range(len(bm)):
        if bm[i]<18.5:
            bmiset[0]+=1
            continue
        elif bm[i]>18.5 and bm[i]<24.9:
            bmiset[1]+=1
            continue
        elif bm[i]>25.0 and bm[i]<29.9:
            bmiset[2]+=1
            continue
        else:
            bmiset[3]+=1
            continue
    return bmiset
        
    
    
```
