```python
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import numpy as np
import math
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
dataset = pd.read_csv("housing.csv")
dataset=dataset.dropna(how='any')
X=dataset[["longitude","latitude","housing_median_age","total_rooms","total_bedrooms","population","households","median_income"]]
y=dataset[["median_house_value"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,  shuffle=True, random_state=1)

bagging_params = {
    "n_estimators" : [50,100,150],
    "max_samples" : [0.25,0.5,1.0],
    "max_features" : [0.25,0.5,1.0],
    "bootstrap" : [True,False],
    "bootstrap_features" : [True,False]
}
rf_params = {
    "n_estimators" : [10,50,100],
    "max_depth" : [3,6,9,14],
    "max_features" : ["auto","sqrt"],
    "criterion" : ["mse","mae"],
    "bootstrap" : [True,False]
}
tree_params = {
    "criterion":["mse", "friedman_mse", "mae", "poisson"],
    "splitter" : ["best","random"],
    "max_depth" : [2,6,10,14],
    "max_features" : ["auto","sqrt","log2",10,None]
}
```

```python
decision_tree_model = DecisionTreeRegressor(max_depth=10) 
tree_model1 = decision_tree_model.fit(X_train,y_train)
predict1 = tree_model1.predict(X_test) # 학습된 의사결정나무 모형으로 평가 데이터 예측
print("RMSE: {}".format(math.sqrt(mean_squared_error(predict1, y_test))))

tree_model_grid=GridSearchCV(DecisionTreeRegressor(max_depth=10),tree_params,
                         scoring="neg_mean_squared_error",
                         verbose = 1,
                         n_jobs = -1)
tree_model_grid.fit(X_train,y_train)

tree_model_grid.best_score_
```
